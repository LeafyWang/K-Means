```python
import time
import math
import random

k = 3
dimension = 2
size = 0

maxDistance = 0
dc = 10 # cutoff distance
```

```python
def calDistance(a,b):
    d = 0
    for i in range(dimension):
        d += (a[i]-b[i])**2
    return math.sqrt(d)


def createData(n):
    cdata = []
    for i in range(n):
        cdata.append([random.randint(1,20),random.randint(1,20)])
        cdata.append([random.randint(10,20),random.randint(30,50)])
        cdata.append([random.randint(40,80),random.randint(20,60)])
    return cdata
```

```python
# possible center points

data = createData(7000)
size = len(data)

timeA= time.time()
distance = [([0] * size) for i in range(size)]
print 123
#print distance

for i in range(size):
    for j in range(i+1,size):
        dtemp = calDistance(data[i],data[j])
        distance[i][j] = dtemp
        distance[j][i] = dtemp
        if dtemp > maxDistance:
            maxDistance = dtemp
#print distance
```

```python
# local density
density = [0 for i in range(size)]

for i in range(size):
    pcount = -1
    for j in range(size):
        if(distance[i][j]<dc):
            pcount += 1
    density[i] = pcount
print "still alive 2" 
#print density
```

```python
# minimum distance between the point and any other point with a higher density
minDistance = [[0,i] for i in range(size)]

for i in range(size):
    minD = maxDistance
    for j in range(size):
        if density[j]>=density[i] and distance[i][j]<minD and j!=i :
            minD = distance[i][j]
    minDistance[i][0] = minD

print "still alive 3"
#print minDistance
            
minDistance.sort(key=lambda d:d[0],reverse=True)
#print minDistance
print "still alive 4"
potential = [data[i[1]] for i in minDistance[:k]]
print potential
print "still alive 5"

timeB = time.time()
print(timeB-timeA)
```
